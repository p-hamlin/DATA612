This assignemnt is simply taking in two inputs from the user and finding the sum of the values inbetween the two input values. If the second value is less than the first, the program will return zero. If the values are the same, that value will be the one printed out. Lastly, when the first number is less than the second, the program will start from the first number, and increment the total value until it has made it all the way to the second number, and the total will be printed. This is first done using a for loop, then demonstrated with a while loop.

```python
#defining and creating the function
def sumFunction(num1, num2):
    total = 0
    #for loop going from the first value to the last one
    for x in range(num1, (num2+1)):
        #update the value of 'total' with the numbers from num1 to num2
        total += x
    #print the total
    print(total)
    
#defining my variables
num1 = 0
num2 = 0

#obtaining inputs from user and converting to an int
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

#calling the functin with the given inputs 
sumFunction(num1, num2)
```

```python
def sumFunction(num1, num2):
    total = 0
    #while loop from the first number to the last number
    while num1 < (num2 + 1):
        #update the value of 'total' with the num1
        total += num1
        #end condition, increment num1 for total to update and so loop isn't infinite
        num1 += 1
    #print the total
    print(total)

#defining my variables
num1 = 0
num2 = 0

#obraining inputs from user and converting to an int
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

sumFunction(num1, num2)
```
